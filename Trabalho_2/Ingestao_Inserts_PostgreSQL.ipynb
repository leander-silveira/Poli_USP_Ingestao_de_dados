INSERT RECLAMACOES SCHEMA _RAW - PYTHON (VS Code) - PostgreSQL (DBeaver)

```python
#INSERT RECLAMACOES SCHEMA _RAW - PYTHON (VS Code) - PostgreSQL (DBeaver)

import pandas as pd
from sqlalchemy import create_engine
import os

# Conectar ao banco de dados
def conectar_banco(host, database, user, password):
    # Cria a URL de conexão
    connection_string = f"postgresql+psycopg2://{user}:{password}@{host}/{database}"
    engine = create_engine(connection_string)
    return engine

# Inserir dados no banco de dados
def inserir_dados(engine, df):
    # Substituir vírgulas por pontos e converter a coluna 'indice' para float
    coluna_float = 'indice'
    if coluna_float in df.columns:
        df[coluna_float] = df[coluna_float].astype(str).str.replace(',', '.')
        df[coluna_float] = pd.to_numeric(df[coluna_float], errors='coerce')

    # Converter colunas específicas para int, se necessário
    colunas_int = [
        'quantidade_total_de_clientes_CCS_e_SCR',
        'quantidade_de_clientes_CCS',
        'quantidade_de_clientes_SCR'
    ]
    for coluna in colunas_int:
        if coluna in df.columns:
            df[coluna] = pd.to_numeric(df[coluna], errors='coerce').fillna(0).astype(int)

    # Converter a coluna 'CNPJ_IF' de float64 para VARCHAR (string)
    coluna_varchar = 'CNPJ_IF'
    if coluna_varchar in df.columns:
        # Converter para string, tratando NaN como string vazia e removendo casas decimais
        df[coluna_varchar] = df[coluna_varchar].fillna(0).astype(int).astype(str)

    # Inserir os dados na tabela "reclamacoes", substituindo se já existirem
    df.to_sql('reclamacoes', engine, if_exists='replace', index=False, schema='_raw')

# Função principal
def main():
    # Configurações do banco de dados
    host = "localhost"
    database = "trabalho_2"
    user = "postgres"
    password = "growdev"

    # Caminho para a pasta com os arquivos CSV
    pasta_arquivos = "C:/Users/growdev/Desktop/Injestao_Poli/Trabalho2/Tabelas tratadas por Python/dados_brutos/reclamacoes"

    # Lista para armazenar os DataFrames
    lista_dfs = []

    # Loop através dos arquivos CSV na pasta
    for arquivo in os.listdir(pasta_arquivos):
        if arquivo.endswith(".csv"):
            caminho_arquivo = os.path.join(pasta_arquivos, arquivo)
            print(f"Lendo o arquivo: {caminho_arquivo}")
            df = pd.read_csv(caminho_arquivo)
            lista_dfs.append(df)

    # Concatenar todos os DataFrames em um único DataFrame
    df_concatenado = pd.concat(lista_dfs, ignore_index=True)

    # Exibir informações antes da conversão
    print("Antes da conversão:")
    print(df_concatenado.info())
    print(df_concatenado.head())

    # Conectar ao banco de dados
    engine = conectar_banco(host, database, user, password)

    # Inserir os dados
    inserir_dados(engine, df_concatenado)

    print("\nDados inseridos com sucesso!")

if __name__ == "__main__":
    main()
```
